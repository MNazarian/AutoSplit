```python
import tensorflow as tf
from tensorflow.keras import layers, models, Input
import pandas as pd
import numpy as np
import os
import logging
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

print("=== CPU-OPTIMIERTES HYBRID-MODELL MIT RESNET50V2 ===")
print(f"TensorFlow version: {tf.__version__}")

# Force CPU usage (deaktiviere GPU falls vorhanden)
os.environ['CUDA_VISIBLE_DEVICES'] = '-1'
tf.config.set_visible_devices([], 'GPU')

# === PFADE DEFINIEREN ===
base_dir = r"C:\Users\meh91075\Desktop\00_all_in_one_dataset"
csv_file = r"C:\Users\meh91075\Desktop\AutoSplit_all_in_one_dataset_20250204.csv"
png_dir = os.path.join(base_dir, 'all_in_one_PNG')

print(f"Base directory: {base_dir}")
print(f"CSV file: {csv_file}")
print(f"PNG directory: {png_dir}")

# === CSV-DATEI LADEN ===
try:
    df = pd.read_csv(csv_file)
    logger.info(f"DataFrame loaded successfully: {df.shape}")
    
    # Entferne label_standard falls vorhanden
    if 'label_standard' in df.columns:
        df = df.drop('label_standard', axis=1)
        logger.info("Dropped 'label_standard' column")
        
except Exception as e:
    logger.error(f"Error loading CSV: {e}")
    raise

# Filter für Klassen 1-4
df_filtered = df[df['label_majority'].isin([1, 2, 3, 4])].copy()
logger.info(f"Filtered for classes 1-4: {len(df_filtered)} rows")
print("Label distribution after filtering:")
print(df_filtered['label_majority'].value_counts().sort_index())

# === BILDER VERARBEITEN ===
def extract_filename(image_path):
    """Extrahiert Dateiname ohne Erweiterung und Suffixe"""
    basename = os.path.basename(image_path)
    filename = basename
    for suffix in ['_ISO', '_ISOAlt', '_iso', '_isoalt']:
        filename = filename.replace(suffix, '')
    filename = os.path.splitext(filename)[0]
    return filename

def get_image_paths(directory):
    """Alle Bildpfade aus Verzeichnis sammeln"""
    if not os.path.exists(directory):
        logger.warning(f"Directory {directory} does not exist")
        return []
    
    image_paths = []
    for file in os.listdir(directory):
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_paths.append(os.path.join(directory, file))
    
    logger.info(f"Found {len(image_paths)} images in {directory}")
    return image_paths

# Bildpfade sammeln
png_images = get_image_paths(png_dir)

# DataFrame mit Bildpfaden erstellen
image_df = pd.DataFrame({
    'image_path': png_images,
    'filename': [extract_filename(path) for path in png_images]
})

# Entferne Duplikate
image_df = image_df.drop_duplicates(subset='filename', keep='first')

# Merge DataFrames
merged_df = pd.merge(df_filtered, image_df[['filename', 'image_path']], on='filename', how='left')
merged_df['has_image'] = merged_df['image_path'].notna()

# Filter nur Einträge mit Bildern
balanced_df = merged_df[merged_df['has_image']].copy()

logger.info(f"Final dataset: {len(balanced_df)} rows with images")
print("Final label distribution:")
print(balanced_df['label_majority'].value_counts().sort_index())

# === TRAIN-TEST SPLIT ===
try:
    train_df, val_df = train_test_split(
        balanced_df, 
        test_size=0.2, 
        stratify=balanced_df['label_majority'], 
        random_state=42
    )
    logger.info(f"Train: {len(train_df)}, Validation: {len(val_df)}")
except Exception as e:
    logger.error(f"Error in train_test_split: {e}")
    raise

# === NUMERISCHE FEATURES ===
numeric_columns = balanced_df.select_dtypes(include=[np.number]).columns.tolist()
numeric_columns = [col for col in numeric_columns if col not in ['label_majority', 'has_image']]
logger.info(f"Numeric features: {len(numeric_columns)}")

# Normalisierung
scaler = StandardScaler()
train_df_scaled = train_df.copy()
val_df_scaled = val_df.copy()

train_df_scaled[numeric_columns] = scaler.fit_transform(train_df[numeric_columns])
val_df_scaled[numeric_columns] = scaler.transform(val_df[numeric_columns])

# === CPU-OPTIMIERTE KONFIGURATION MIT RESNET50V2 ===
img_height, img_width = 224, 224
cpu_batch_size = 8       # Reduziert wegen ResNet50V2
cpu_epochs = 100         # Moderater Anzahl Epochen
initial_learning_rate = 1e-3

logger.info(f"CPU Configuration with ResNet50V2: Batch Size={cpu_batch_size}, Epochs={cpu_epochs}")

# === CPU-OPTIMIERTE DATASET-FUNKTION MIT RESNET-PREPROCESSING ===
def create_cpu_dataset_with_resnet_preprocessing(dataframe, batch_size, is_training=True):
    """CPU-optimierte Dataset-Funktion mit ResNet-Preprocessing"""
    
    def parse_function(filename, label, numeric_data):
        # Bild laden und für ResNet50V2 preprocessen
        try:
            img = tf.io.read_file(filename)
            img = tf.image.decode_png(img, channels=3)
            img = tf.image.resize(img, [img_height, img_width])
            
            # ResNet50V2 Preprocessing (wichtig für pretrained model)
            img = tf.keras.applications.resnet_v2.preprocess_input(img)
            
            # Minimale Augmentation für CPU (weniger rechenintensiv)
            if is_training:
                img = tf.image.random_flip_left_right(img)
                img = tf.image.random_brightness(img, max_delta=0.2)
                
        except Exception:
            # Fallback bei Fehlern - ResNet-kompatible Null-Matrix
            img = tf.zeros([img_height, img_width, 3], dtype=tf.float32)
        
        # Numerische Daten verarbeiten
        numeric_data = tf.cast(numeric_data, tf.float32)
        numeric_data = tf.ensure_shape(numeric_data, [len(numeric_columns)])
        
        # Labels anpassen (1-4 → 0-3)
        label = tf.cast(label - 1, tf.int32)
        
        return (img, numeric_data), label
    
    # Dataset erstellen
    numeric_array = dataframe[numeric_columns].values.astype(np.float32)
    
    dataset = tf.data.Dataset.from_tensor_slices((
        dataframe['image_path'].values,
        dataframe['label_majority'].values.astype(np.int32),
        numeric_array
    ))
    
    # CPU-optimierte Pipeline (reduzierte Parallelität wegen ResNet)
    dataset = dataset.map(parse_function, num_parallel_calls=2)  # Weniger parallel für CPU
    
    if is_training:
        dataset = dataset.shuffle(buffer_size=500).repeat()  # Kleinerer Buffer
    
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(1)  # Kleinerer Buffer für CPU
    
    return dataset

# Datasets erstellen
logger.info("Creating CPU-optimized datasets with ResNet preprocessing...")
cpu_train_dataset = create_cpu_dataset_with_resnet_preprocessing(train_df_scaled, cpu_batch_size, is_training=True)
cpu_val_dataset = create_cpu_dataset_with_resnet_preprocessing(val_df_scaled, cpu_batch_size, is_training=False)

# Schritte berechnen
cpu_steps_per_epoch = max(1, len(train_df) // cpu_batch_size)
cpu_validation_steps = max(1, len(val_df) // cpu_batch_size)

logger.info(f"Steps per epoch: {cpu_steps_per_epoch}, Validation steps: {cpu_validation_steps}")

# Dataset-Test
try:
    for (img_batch, num_batch), label_batch in cpu_train_dataset.take(1):
        logger.info(f"Dataset test successful!")
        logger.info(f"Image batch shape: {img_batch.shape}, Numeric batch shape: {num_batch.shape}")
        break
except Exception as e:
    logger.error(f"Dataset test failed: {e}")
    raise

# === CPU-OPTIMIERTES HYBRID-MODELL MIT RESNET50V2 ===
def create_cpu_hybrid_model_with_resnet(img_height, img_width, num_numeric_features, num_classes):
    """CPU-optimiertes Hybrid-Modell mit vortrainiertem ResNet50V2"""
    
    # Bild-Input (CNN mit ResNet50V2)
    img_input = Input(shape=(img_height, img_width, 3), name='image_input')
    
    # Verwende vortrainiertes ResNet50V2 (Transfer Learning)
    base_model = tf.keras.applications.ResNet50V2(
        include_top=False, 
        weights='imagenet', 
        input_tensor=img_input
    )
    
    # Freeze die vortrainierten Layer für CPU-Optimierung
    base_model.trainable = False
    
    # Feature-Extraktion aus ResNet50V2
    x = base_model.output
    x = layers.GlobalAveragePooling2D()(x)
    
    # Reduzierte Dense Layer für CPU-Performance
    x = layers.Dense(256, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)
    
    # Numerischer Input (MLP) - optimiert für CPU
    num_input = Input(shape=(num_numeric_features,), name='numeric_input')
    y = layers.Dense(128, activation='relu')(num_input)
    y = layers.BatchNormalization()(y)
    y = layers.Dropout(0.3)(y)
    y = layers.Dense(64, activation='relu')(y)
    y = layers.BatchNormalization()(y)
    y = layers.Dropout(0.2)(y)
    
    # Combine branches
    combined = layers.concatenate([x, y])
    
    # Final layers (reduziert für CPU)
    z = layers.Dense(128, activation='relu')(combined)
    z = layers.BatchNormalization()(z)
    z = layers.Dropout(0.5)(z)
    z = layers.Dense(64, activation='relu')(z)
    z = layers.BatchNormalization()(z)
    z = layers.Dropout(0.3)(z)
    
    # Output layer
    output = layers.Dense(num_classes, activation='softmax')(z)
    
    model = models.Model(inputs=[img_input, num_input], outputs=output)
    return model

# Modell erstellen
logger.info("Creating CPU-optimized hybrid model with ResNet50V2...")
num_classes = 4

cpu_model = create_cpu_hybrid_model_with_resnet(img_height, img_width, len(numeric_columns), num_classes)

cpu_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=initial_learning_rate),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("CPU-Model with ResNet50V2 Summary:")
cpu_model.summary()

# === CALLBACKS ===
cpu_callbacks = [
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss', 
        factor=0.2, 
        patience=3,  
        min_lr=1e-6, 
        verbose=1
    ),
    tf.keras.callbacks.ModelCheckpoint(
        os.path.join(base_dir, 'best_cpu_resnet_hybrid_model.h5'),
        save_best_only=True,
        monitor='val_loss',
        verbose=1
    ),
    tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=7,  # Frühheres Stoppen für CPU
        restore_best_weights=True,
        verbose=1
    )
]

# === TRAINING STARTEN ===
print("\n" + "="*60)
print("🚀 STARTE CPU-OPTIMIERTES TRAINING MIT RESNET50V2")
print("="*60)
print("🔥 ResNet50V2 auf CPU - Training dauert deutlich länger!")
print("📈 Erwartete CPU-Auslastung: 80-100%")
print("💾 Erwarteter RAM-Verbrauch: 16-32GB")
print("🕐 Geschätzte Dauer: 6-12 Stunden")
print("🎯 Bessere Accuracy durch Transfer Learning erwartet")

try:
    logger.info("Starting CPU-optimized training with ResNet50V2...")
    cpu_history = cpu_model.fit(
        cpu_train_dataset,
        epochs=cpu_epochs,
        steps_per_epoch=cpu_steps_per_epoch,
        validation_data=cpu_val_dataset,
        validation_steps=cpu_validation_steps,
        callbacks=cpu_callbacks,
        verbose=1
    )
    logger.info("Training completed successfully!")
    
except Exception as e:
    logger.error(f"Training failed: {e}")
    raise

# === EVALUATION ===
print("\n" + "="*60)
print("📊 EVALUATION")
print("="*60)

# Lade beste Gewichte
try:
    cpu_model.load_weights(os.path.join(base_dir, 'best_cpu_resnet_hybrid_model.h5'))
    logger.info("Best weights loaded")
except:
    logger.warning("Could not load best weights, using current weights")

# Evaluation
logger.info("Evaluating model...")
val_metrics = cpu_model.evaluate(cpu_val_dataset, steps=cpu_validation_steps, verbose=1)
logger.info(f"Validation Loss: {val_metrics[0]:.4f}")
logger.info(f"Validation Accuracy: {val_metrics[1]:.4f}")

# === PREDICTIONS ===
logger.info("Generating predictions...")
y_pred_probs = cpu_model.predict(cpu_val_dataset, steps=cpu_validation_steps, verbose=1)
y_pred_classes = np.argmax(y_pred_probs, axis=1)

# True labels sammeln
y_true = []
for i, (_, y_batch) in enumerate(cpu_val_dataset):
    if i >= cpu_validation_steps:
        break
    y_true.extend(y_batch.numpy())
y_true = np.array(y_true[:len(y_pred_classes)])

# === METRIKEN BERECHNEN ===
accuracy = accuracy_score(y_true, y_pred_classes)
precision_macro = precision_score(y_true, y_pred_classes, average='macro')
recall_macro = recall_score(y_true, y_pred_classes, average='macro')
f1_macro = f1_score(y_true, y_pred_classes, average='macro')

precision_weighted = precision_score(y_true, y_pred_classes, average='weighted')
recall_weighted = recall_score(y_true, y_pred_classes, average='weighted')
f1_weighted = f1_score(y_true, y_pred_classes, average='weighted')

conf_matrix = confusion_matrix(y_true, y_pred_classes)
class_report = classification_report(
    y_true, y_pred_classes, 
    target_names=[f'Class {i+1}' for i in range(4)],
    digits=4
)

# === ERGEBNISSE AUSGEBEN ===
print("\n" + "="*60)
print("🎯 FINALE ERGEBNISSE - RESNET50V2 AUF CPU")
print("="*60)
print(f"Final Validation Loss: {val_metrics[0]:.4f}")
print(f"Final Validation Accuracy: {val_metrics[1]:.4f}")
print(f"Accuracy from predictions: {accuracy:.4f}")

print(f"\n📊 MACRO AVERAGE:")
print(f"  Precision: {precision_macro:.4f}")
print(f"  Recall: {recall_macro:.4f}")
print(f"  F1-Score: {f1_macro:.4f}")

print(f"\n📊 WEIGHTED AVERAGE:")
print(f"  Precision: {precision_weighted:.4f}")
print(f"  Recall: {recall_weighted:.4f}")
print(f"  F1-Score: {f1_weighted:.4f}")

print(f"\n📈 CONFUSION MATRIX:")
print(conf_matrix)

print(f"\n📋 CLASSIFICATION REPORT:")
print(class_report)

# === PER-CLASS PERFORMANCE ===
print(f"\n🎯 PER-CLASS PERFORMANCE:")
for i in range(4):
    class_accuracy = np.sum((y_true == i) & (y_pred_classes == i)) / np.sum(y_true == i)
    class_count = np.sum(y_true == i)
    print(f"  Class {i+1}: {class_accuracy:.1%} (n={class_count})")

# === VISUALISIERUNGEN ===
print("\n📈 Creating visualizations...")

# Training History
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(cpu_history.history['accuracy'], label='Training Accuracy', linewidth=2)
plt.plot(cpu_history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
plt.title('Model Accuracy (CPU Training with ResNet50V2)', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(cpu_history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(cpu_history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('Model Loss (CPU Training with ResNet50V2)', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(base_dir, 'cpu_resnet_training_history.png'), dpi=300, bbox_inches='tight')
plt.show()

# Confusion Matrix Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Class 1', 'Class 2', 'Class 3', 'Class 4'],
            yticklabels=['Class 1', 'Class 2', 'Class 3', 'Class 4'],
            cbar_kws={'label': 'Count'})
plt.title('Confusion Matrix (CPU Training with ResNet50V2)', fontsize=16, fontweight='bold')
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.savefig(os.path.join(base_dir, 'cpu_resnet_confusion_matrix.png'), dpi=300, bbox_inches='tight')
plt.show()

# Learning Rate Plot (falls im History)
if 'lr' in cpu_history.history:
    plt.figure(figsize=(10, 6))
    plt.plot(cpu_history.history['lr'], linewidth=2)
    plt.title('Learning Rate Schedule (CPU Training with ResNet50V2)', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.yscale('log')
    plt.grid(True, alpha=0.3)
    plt.savefig(os.path.join(base_dir, 'cpu_resnet_learning_rate.png'), dpi=300, bbox_inches='tight')
    plt.show()

# === MODELL UND ERGEBNISSE SPEICHERN ===
final_model_path = os.path.join(base_dir, 'final_cpu_resnet_hybrid_model.h5')
cpu_model.save(final_model_path)
logger.info(f"Final model saved to: {final_model_path}")

# Model architecture visualization
try:
    tf.keras.utils.plot_model(
        cpu_model, 
        to_file=os.path.join(base_dir, 'cpu_resnet_model_architecture.png'), 
        show_shapes=True, 
        show_layer_names=True
    )
    logger.info("Model architecture saved")
except:
    logger.warning("Could not save model architecture plot")

# Predictions speichern
predictions_df = pd.DataFrame({
    'True_Label': y_true + 1,  # Zurück zu 1-4
    'Predicted_Class': y_pred_classes + 1,  # Zurück zu 1-4  
    'Confidence': np.max(y_pred_probs, axis=1),
    'Class_1_Prob': y_pred_probs[:, 0],
    'Class_2_Prob': y_pred_probs[:, 1],
    'Class_3_Prob': y_pred_probs[:, 2],
    'Class_4_Prob': y_pred_probs[:, 3]
})
predictions_df.to_csv(os.path.join(base_dir, 'cpu_resnet_predictions.csv'), index=False)

# Zusammenfassung speichern
summary_content = f"""=== CPU HYBRID MODEL WITH RESNET50V2 TRAINING SUMMARY ===

Dataset Information:
- Total samples with images: {len(balanced_df)}
- Training samples: {len(train_df)}
- Validation samples: {len(val_df)}
- Numeric features: {len(numeric_columns)}

Model Configuration:
- Architecture: CPU-optimized Hybrid with ResNet50V2 (CNN + MLP)
- Base Model: ResNet50V2 (pretrained on ImageNet, frozen)
- Image input: {img_height}x{img_width}x3
- Numeric input: {len(numeric_columns)} features
- Output classes: {num_classes}
- Total parameters: {cpu_model.count_params():,}
- ResNet50V2 frozen: True (for CPU optimization)

Training Configuration:
- Batch size: {cpu_batch_size} (reduced for ResNet50V2)
- Training epochs: {len(cpu_history.history['loss'])}
- Steps per epoch: {cpu_steps_per_epoch}
- Validation steps: {cpu_validation_steps}
- Initial learning rate: {initial_learning_rate}

Performance Metrics:
- Final Validation Loss: {val_metrics[0]:.4f}
- Final Validation Accuracy: {val_metrics[1]:.4f}

Macro Average:
- Precision: {precision_macro:.4f}
- Recall: {recall_macro:.4f}
- F1-Score: {f1_macro:.4f}

Weighted Average:
- Precision: {precision_weighted:.4f}
- Recall: {recall_weighted:.4f}
- F1-Score: {f1_weighted:.4f}

Confusion Matrix:
{conf_matrix}

Per-Class Performance:
{chr(10).join([f"Class {i+1}: {np.sum((y_true == i) & (y_pred_classes == i)) / np.sum(y_true == i):.1%} (n={np.sum(y_true == i)})" for i in range(4)])}

Transfer Learning Benefits:
- Pretrained ResNet50V2 features
- ImageNet initialization
- Better feature extraction expected

Files Saved:
- Model: {final_model_path}
- Best weights: best_cpu_resnet_hybrid_model.h5
- Predictions: cpu_resnet_predictions.csv
- Training history: cpu_resnet_training_history.png
- Confusion matrix: cpu_resnet_confusion_matrix.png
- Model architecture: cpu_resnet_model_architecture.png
"""

with open(os.path.join(base_dir, 'cpu_resnet_training_summary.txt'), 'w') as f:
    f.write(summary_content)

# Analyse von misclassified examples
misclassified = np.where(y_pred_classes != y_true)[0]
print(f"\nNumber of misclassified examples: {len(misclassified)}")

if len(misclassified) > 0:
    print("\nDetails of some misclassified examples:")
    for i in misclassified[:5]:  # Show details for the first 5 misclassified examples
        true_label = y_true[i]
        pred_label = y_pred_classes[i]
        pred_probs = y_pred_probs[i] # Probabilities for all classes
        print(f"Example {i}: True Label: {true_label+1}, Predicted Label: {pred_label+1}")
        print(f"  Confidence: {np.max(pred_probs):.3f}")
        print(f"  All probabilities: {[f'{p:.3f}' for p in pred_probs]}")

# === ABSCHLUSS ===
print("\n" + "="*60)
print("🎉 CPU TRAINING MIT RESNET50V2 ERFOLGREICH ABGESCHLOSSEN!")
print("="*60)
print(f"✅ Final accuracy: {val_metrics[1]:.1%}")
print(f"✅ Total epochs: {len(cpu_history.history['loss'])}")
print(f"✅ Model saved: {final_model_path}")
print(f"✅ Results saved to: {base_dir}")

print(f"\n🔥 RESNET50V2 PERFORMANCE:")
print("✅ Transfer Learning erfolgreich implementiert")
print("✅ Pretrained ImageNet features genutzt")
print("✅ Gefrorene ResNet-Layer für CPU-Optimierung")

print(f"\n💡 VERBESSERUNGSVORSCHLÄGE:")
print("1. 🖥️  GPU-Setup für 10x Beschleunigung")
print("2. 🔓 Fine-tuning: ResNet-Layer entfrieren")
print("3. 🌐 Cloud-GPU für bessere Performance") 
print("4. 🔄 Weitere Data Augmentation")

print(f"\n🎯 NÄCHSTE SCHRITTE:")
print("1. Vergleichen Sie Ergebnisse mit einfacher CNN")
print("2. Analysieren Sie die Confusion Matrix")
print("3. Untersuchen Sie misklassifizierte Beispiele")
print("4. Für Produktion: GPU-Version implementieren")

logger.info("CPU training with ResNet50V2 completed successfully!")

print("\n" + "="*60)
print("📄 ALLE ERGEBNISSE GESPEICHERT:")
print("="*60)
print(f"• Training Summary: cpu_resnet_training_summary.txt")
print(f"• Model: final_cpu_resnet_hybrid_model.h5")
print(f"• Best Weights: best_cpu_resnet_hybrid_model.h5")  
print(f"• Predictions: cpu_resnet_predictions.csv")
print(f"• Training Plots: cpu_resnet_training_history.png")
print(f"• Confusion Matrix: cpu_resnet_confusion_matrix.png")
print(f"• Architecture: cpu_resnet_model_architecture.png")
```
